```python
df=df.drop(['brand','model','base_color','processor','screen_size'],axis=1)
```

```python
# Correlation Analysis between features
plt.figure(figsize=(12,10))
corr_matrix = df.corr()
sns.heatmap(corr_matrix, cmap = 'YlGnBu', annot = True, fmt = '.2f')
```

```python
corr_threshold = 0.7
high_corr_features = set()
for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if abs(corr_matrix.iloc[i, j]) > corr_threshold:
            colname = corr_matrix.columns[i]
            high_corr_features.add(colname)

df_reduced = df.drop(high_corr_features, axis = 1)

print("Original dataset shape:", df.shape)
print("Reduced dataset shape:", df_reduced.shape)
```

```python
x = df.iloc[:,0:10]
y = df.iloc[:,10]

print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
import sklearn
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)

x_train = pd.DataFrame(x_train)
x_train.head()
```

```python
import tensorflow as tf
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=[10]),
    tf.keras.layers.Dense(64, activation='relu', input_shape=[16]),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(1, activation='linear')
])
```

```python
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['mean_absolute_error'])
```

```python
history = model.fit(x_train, y_train, epochs=100, validation_data=(x_test, y_test))
```

```python
# Visualize the performance of the ANN model
plt.plot(history.history['loss'], label='training loss')
plt.plot(history.history['val_loss'], label='validation loss')
plt.legend()
plt.show()
```

```python
plt.plot(history.history['mean_absolute_error'], label='training MAE')
plt.plot(history.history['val_mean_absolute_error'], label='validation MAE')
plt.legend()
plt.show()
```

```python
loss, accuracy = model.evaluate(x_test, y_test)
print(f'Test loss: {loss}, Test accuracy: {accuracy}')
```
